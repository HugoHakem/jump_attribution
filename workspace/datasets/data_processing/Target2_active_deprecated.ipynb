# Target2 Compound Filtering for an Unbiased and Phenotypically Active Dataset  

## Dataset Seletion
We use the JUMP Cell Painting datasets (*Chandrasekaran et al., 2023*²), which are available from the Cell Painting Gallery (*Weisbart et al., 2024*³) on the [Registry of Open Data on AWS](https://registry.opendata.aws/cellpainting-gallery/). This is the largest open source Cell Painting dataset to date (2024). 

The framework we propose is computationnally intensive. For that reason, subsetting the dataset into a smaller one is key. We decided relying on:
* [JUMP-Target-2-Compound](https://github.com/jump-cellpainting/JUMP-Target) datasets. Those compound are well annotated and Mechanism of Action (MoA) information is well available.
* This dataset is composed of 301 compounds in total, of which 46 are included as controls serving different goals.
* The plate layout is well normalized in this dataset. There are:
   -  n=2 replicate wells for each of 14 poscon_diverse compounds
   -  n=64 DMSO wells
   -  n=1 well for the remaining compounds.
  
  To avoid any bias relative to source or plate layout, we suggest using all Target-2-Compound, but also merging with the [compound dataset](https://github.com/jump-cellpainting/datasets). Doing so, we keep well annotated compound but enlarge to larger plate layout and source provenance.

## Phenotypic active compound
In our previous approaches the classifier’s performance on cell-images was suboptimal. We suspect those results were due to some compound or replicates being not phenotypically active -- Hence no any different from DMSO. Since our interpretation framework requires a well-performing classifier, we decided refining the downsampling strategy for the Target2 dataset. 

To achieve this, we propose using mean Average Precision (mAP) to retain only replicates that are both phenotypically active and consistent within their mechanism of action (MoA). mAP will be computed using the `copairs` package, as described in *Kalinin et al., 2024*¹.

## Unbiased metadata filtering
Once phenotypically active compound are retrieved. We want to diminish the risk of the classification being occuring on well or source position. We suggest downsampling the dataset so that it is well balance on source, plate and well position.

```python
# Removal of source_9 and source_1 because the plate is not similar
well = well.filter(pl.col("Metadata_Source").str.contains("_9|_1$") != True)

plate = plate.select(pl.all().exclude("Metadata_Source"))

micro_features = ['Metadata_Source','Metadata_Microscope_Name',
                  'Metadata_Widefield_vs_Confocal',
                  'Metadata_Excitation_Type',
                  'Metadata_Objective_NA',
                  'Metadata_Filter_Configuration']

#rename source so it is consistent with the other data
micro = micro.with_columns(
    ("source_" + pl.col("Metadata_Source").cast(str)).alias("Metadata_Source")).select(
    pl.col(micro_features))

#create a unique ID identifier per microscope configuration
micro = micro.filter(pl.col("Metadata_Source").str.contains("_[19]$|_15$") != True)
map_micro_ID = micro.select(pl.col("Metadata_Source"),
                            pl.struct(pl.all().exclude("Metadata_Source")).alias("unique"))\
                            .sort(by="unique")\
                            .select(pl.col("Metadata_Source"), pl.col("unique").rle_id().alias("Micro_id"))

micro = micro.join(map_micro_ID,
           on="Metadata_Source",
           how="left")
```

```python
# check if InChIKey is a unique identifier
target2.group_by("InChIKey").agg(pl.all().n_unique()).with_columns(
    pl.sum_horizontal(pl.all().exclude("InChIKey"))).filter(pl.col("sum")>8)
```
